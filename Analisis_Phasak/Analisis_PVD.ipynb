```python
# importar librerías
import pandas as pd
import numpy as np
```

```python
# Obtenermos datos de CSV de españa
df = pd.read_csv("./Articulos.csv",sep=";",encoding="latin")
```

```python
# Renombramos alguna columna
df = df.rename(columns={"Precio coste medio":"Coste Medio"})
```

```python
df.head(15)
```

```python
# Comprobamos si hay datos nulos
df.isna().sum()
```

```python
# Reemplazamos las , por . en los decimales.
df["Coste Medio"] = df["Coste Medio"].str.replace(',','.')
```

```python
df.head(15)
```

```python
# Pasamos el Coste medio a decimal
df["Coste Medio"] = df["Coste Medio"].astype('float')
```
