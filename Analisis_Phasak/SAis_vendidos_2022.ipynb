```python
def impcsv():
    df['Precio'] = df['Precio'].str.replace(',','.')
    df['Importe'] = df['Importe'].str.replace(',','.')
    df['Precio medio'] = df['Precio medio'].str.replace(',','.')
    df['Precio'] = df['Precio'].astype('float64')
    df['Importe'] = df['Importe'].astype('float64')
    df['Precio medio'] = df['Precio medio'].astype('float64')
    df['Codigo Cliente'] = df['Codigo Cliente'].astype('object')
    df['Serie Factura'] = df['Serie Factura'].astype('object')
    df['Codigo Comercial'] = df['Codigo Comercial'].astype('object')
    df['Empresa Emisora'] = df['Empresa Emisora'].astype('object')
    return
```

## Diferentes formas de hacer un filtro

### La Forma Larga:
### Apoyandote en una lista con el mismo numero de elementos que filas tiene el dataframe

---

```python
df.shape
```

```python
#Creamos una lista, recorremos la Serie del dataFrame añadiendo True o False en cada una
booleans = []
for importe in df.Importe:
    if importe  >= 300.0:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
# Como vemos la longitud de la lista es igual a la del DataFrame
len(booleans)
```

```python
#Convertimos la lista de booleans en una Serie de Pandas
```

```python
tope_importe = pd.Series(booleans)
```

```python
tope_importe.head()
```

```python
# Le pasamos la variable al dataframe con corchetes y nos muestra los que sumplen la condicion. 
# en este caso el importe mayor a 300.0
df[tope_importe]
```

---

### Esta es otra forma mas corta

```python
# Hacemos el filtro sobre la serie del DataFrame y devuleve una serie de True y False que lo guardamos en una variable 
tope_importe = df.Importe >= 300.0
tope_importe.head()
```

```python
df[tope_importe]
```

---

### Y la ultima y mas rapida forma de hacer el filtro

```python
df[df.Importe >= 300.0]
```

```python
# Aqui mostrando solo la columna que elegimos. Ya filtrada. Aqui sin corchetes pero separado con un punto
df[df.Importe >= 300.0].Importe
```

```python
# Aqui tambien solo la columna elegida. Ya filtra. Aqui con corchetes y sin poner el punto de separacion.
df[df.Importe >= 300.0]['Importe']
```

### Y esta es la mejor forma de hacerlo con loc.

```python
# Separado por una coma. En la primera parte indicamos que 
# filas mostar, y en la segunda que columnas mostrar.
df.loc[df.Importe >= 300.0, 'Importe']
```
