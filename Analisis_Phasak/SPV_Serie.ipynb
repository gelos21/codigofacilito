```python
dfsn = pd.read_csv('Numeros de serie sais 18-22.csv', encoding='utf8', sep=';')
```

```python
dfsn.rename(columns={'Mov Alm -> Referencia': 'Referencia', 'Número de Serie': 'SN', 'Mov Alm -> Fecha': 'Fecha',
       'Mov Alm -> Salidas': 'Salidas', 'Mov Alm -> Entradas': 'Entradas',
       'Mov Alm -> Cantidad con signo': 'Cantidad_Signo'}, inplace=True)
```

```python
dfsn.head()
```

```python
dfsn.Fecha.replace(to_replace=r'Ene', value='Jan', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Feb', value='Feb', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Mar', value='Mar', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Abr', value='Apr', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'May', value='May', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Jun', value='Jun', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Jul', value='Jul', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Ago', value='Aug', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Sep', value='Sep', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Oct', value='Oct', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Nov', value='Nov', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'Dic', value='Dec', regex=True, inplace=True)
```

```python
dfsn.Fecha.replace(to_replace=r'-Ene-', value='/01/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Feb-', value='/02/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Mar-', value='/03/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Abr-', value='/04/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-May-', value='/05/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Jun-', value='/06/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Jul-', value='/07/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Ago-', value='/08/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Sep-', value='/09/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Oct-', value='/10/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Nov-', value='/11/', regex=True, inplace=True)
dfsn.Fecha.replace(to_replace=r'-Dic-', value='/12/', regex=True, inplace=True)
```
