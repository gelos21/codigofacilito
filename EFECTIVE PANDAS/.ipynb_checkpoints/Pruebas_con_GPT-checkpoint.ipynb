```python
df['year'] = df['fecha'].dt.year
```

```python
df.head()
```

```python
df.dtypes
```

```python
df_grouped_year = df.groupby(['year','mes','referencia'])
```

```python
df_grouped_year.head()
```

```python
df_grouped_salidas = df_grouped_year['salidas'].sum()
```

```python
df_grouped_salidas.head(20)
```

```python
df_result = df_grouped_salidas.reset_index()
```

```python
df_result
```

```python
df_result.dtypes
```

```python
filter_result_2022 = df_result['year']== 2022 # creamos el filtro de year igual a 2022. Devuelve booleanos.
```

```python
df_result_2022 = df_result[filter_result_2022] # Pasamos el filtro al dataframe para que nos de datos y los guardamos
```

```python
df_result_2022 # Aqui ya tenemos agrupados por year-mes-referencia sumado por salidas y filtrado por year 2022
```

```python
df_pivoted = df_result_2022.pivot_table(index='referencia', columns='mes', values='salidas')
```
