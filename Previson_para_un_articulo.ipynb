```python
# Importar librerías
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Cargar datos en un DataFrame
df = pd.read_csv("/home/ubuntu/projects/codigofacilito/sais_definitivo.csv", sep=";")

# Seleccionar predictoras y variable a predecir
X = df[["referencia", "mes"]] # predictoras
y = df["salidas"] # variable a predecir

# Crear un objeto LabelEncoder
label_encoder = LabelEncoder()

# Transformar la columna "referencia" a valores numéricos
X.loc[:, "referencia"] = label_encoder.fit_transform(X["referencia"])

# Asignar nombres a las características
X.columns = ["referencia_num", "mes"]

# Crear un modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X, y)

# Hacer predicciones sobre nuevos datos
predictions = model.predict([[9100, 1], [9100, 2], [9100, 3]])

# Mostrar predicciones
print(predictions)
```
